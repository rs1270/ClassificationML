```python
import pandas as pd
df2=pd.read_csv('GUIDE_Test.csv')
```

```python
df2.head()
```

```python
df2.info()
```

```python
(df2.isnull().sum()/(len(df2)))*100
threshold= len(df2)* 0.5   # removing 50% above null value
df2=df2.dropna(thresh=threshold,axis=1)
```

```python
df2.isnull().sum()
```

```python
df2['Timestamp']=pd.to_datetime(df2['Timestamp'])
df2["Day"]=df2["Timestamp"].dt.day
df2["Month"]=df2["Timestamp"].dt.month
df2["Year"]=df2["Timestamp"].dt.year
df2["Hour"]=df2["Timestamp"].dt.hour
df2.drop("Timestamp",axis=1,inplace=True)
```

```python
category_percentages = (df2['IncidentGrade'].value_counts(normalize=True) * 100).round(2)


print(category_percentages)
```

```python
print(df2.info())
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
numerical_cols = df2.select_dtypes(include=['int64', 'float64','int32']).columns

# Initialize the scaler
scaler = MinMaxScaler()

# Normalize each numerical column using a loop
for col in numerical_cols:
    df2[col] = scaler.fit_transform(df2[[col]])
```

```python
freq_encoding = df2['Category'].value_counts(normalize=False)  # Absolute frequency
df2['Category'] = df2['Category'].map(freq_encoding)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df2['IncidentGrade'] = encoder.fit_transform(df2['IncidentGrade'])
```

```python
freq_encoding = df2['EntityType'].value_counts(normalize=False)  # Absolute frequency
df2['EntityType'] = df2['EntityType'].map(freq_encoding)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df2['EvidenceRole'] = encoder.fit_transform(df2['EvidenceRole'])
```

```python
import joblib
import xgboost as xgb
model=joblib.load('model.pkl')
x_test=df2[['NetworkMessageId', 'State', 'IncidentId', 'IpAddress', 'Day', 'EvidenceRole', 
                     'FolderPath', 'RegistryKey', 'RegistryValueName', 'EntityType', 'AccountName', 
                     'Url', 'FileName', 'OSFamily', 'OrgId', 'DeviceId', 'City', 'OSVersion', 'AccountSid', 
                     'ApplicationId', 'ApplicationName', 'Sha256', 'AccountObjectId', 'Id', 'Category', 
                     'AlertId', 'RegistryValueData', 'CountryCode', 'DetectorId', 'Month', 'DeviceName', 
                     'Hour', 'AccountUpn', 'AlertTitle']]


prediction=model.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
y_test=df2['IncidentGrade']
result=confusion_matrix(y_test,prediction)
print('confusion mat')
print(result)
result1 = classification_report(y_test, prediction)
print("Classification Report:",)
print (result1)
```
